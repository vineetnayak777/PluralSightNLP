```python
#Encapsulating all the parsing logic
def getTextWaPo(url):
    #Downloding the Webpage
    page = urllib.request.urlopen(articleURL).read().decode('utf8','ignore')
    #Parsing the data
    soup = BeautifulSoup(page,"lxml")
    #to find all the element with the tag article
    text = ' '.join(map(lambda p: p.text, soup.find_all('article')))
    return text.encode('ascii',errors='replace').decode().replace("?"," ")
```

```python
text = getTextWaPo(articleURL)
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from string import punctuation
```

```python
sents = sent_tokenize(text)
sents
```

```python
word_sent = word_tokenize(text.lower())
word_sent
```

```python
_stopwords = set(stopwords.words('english')+list(punctuation))
_stopwords
```

```python
word_sent = [word for word in word_sent if word not in _stopwords]
```

```python
word_sent
```

```python
#Auto Summarize text by extracting the most important sentences
from nltk.probability import FreqDist #Construct a frequency distribution
freq = FreqDist(word_sent)
freq
```

```python
from heapq import nlargest
```

```python
nlargest(10, freq, key=freq.get)
```

```python
from collections import defaultdict
#Key -> Sentences Value -> Significant Score
```

```python
ranking = defaultdict(int)
#Enumarate converts the list into tuples

for i, j in enumerate(sents):
    for w in word_tokenize(j.lower()):
        if w in freq:
            ranking[i] += freq[w]
            
```

```python
def summarize(text, n):
    sents = sent_tokenize(text)
    
    assert n <= len(sents)
    word_sent = word_tokenize(text.lower())
    _stopwords = set(stopwords.words('english')+list(punctuation))
    
    word_sent = [word for word in word_sent if word not in _stopwords]
    freq = FreqDist(word_sent)
    
    ranking = defaultdict(int)
    #Enumarate converts the list into tuples

    for i, j in enumerate(sents):
        for w in word_tokenize(j.lower()):
            if w in freq:
                ranking[i] += freq[w]
    sents_idx = nlargest(n, ranking, key = ranking.get)
    return [sents[j] for j in sorted(sents_idx)]
```

```python
summarize(text,2)
```
